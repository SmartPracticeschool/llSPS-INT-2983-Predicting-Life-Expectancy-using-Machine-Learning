```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df_data_2 = df_data_1.dropna()
# y = df_data_2["Life expectancy "]
# x_cols = [x for x in df_data_2.columns if x != 'Life expectancy ' and x!='Status' and x!='Country']
# x = df_data_2[x_cols]
# df = df_data_1.loc[:, df_data_1.columns != 'Life expectancy ']
# print(df)
```

```python
dummies = pd.get_dummies(df_data_2['Status'])
dummies.tail()
merged = pd.concat([df_data_2,dummies],axis = 'columns')
final = merged.drop(['Status'],axis = 'columns')
print(final.columns)
data = pd.DataFrame(final)
```

```python
y = data["Life expectancy "]
x_cols = [d for d in data.columns if d != 'Life expectancy 'and d !='Country']
print(x_cols)
x = data[x_cols]
print(x.shape)
```

```python
linear_reg = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
linear_reg.fit(x_train,y_train)
```

```python
y_pred = linear_reg.predict(x_test)
print(linear_reg.score(x_test,y_test))
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```
